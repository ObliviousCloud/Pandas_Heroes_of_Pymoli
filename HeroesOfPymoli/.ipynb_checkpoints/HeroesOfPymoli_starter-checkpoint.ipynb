```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Raw data file
file = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head()
```

```python
#Number of Unique Items
#Average Purchase Price
#Total Number of Purchases
#Total Revenue

unique_items = len(purchase_data["Item ID"].unique())
average_price = round(purchase_data["Price"].mean())
total_purchases = purchase_data["Price"].count()
total_revenue = purchase_data["Price"].sum()

purchasing_analysis = pd.DataFrame({"Number of Uniques Items": [unique_items],
                                   "Average Purchase Price": [average_price],
                                   "Total Purchases": [total_purchases],
                                   "Total Revenue": [total_revenue]
                                   })
purchasing_analysis
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
male = purchase_data.loc[purchase_data['Gender'] == 'Male', ['SN']]
male_players = len(male.groupby(['SN']))

female = purchase_data.loc[purchase_data['Gender'] == 'Female', ['SN']]
female_players = len(female.groupby(['SN']))

other = purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed',['SN']]
other_players = len(other.groupby(['SN']))

Genders = pd.DataFrame(
    {"Percentage of Players": [round((male_players/total_players)*100),
                   round((female_players/total_players)*100),
                   round((other_players/total_players)*100)],
     "Gender": ['Male','Female','Other / Non-Disclosed'],
    "Count of Gender": [male_players, female_players, other_players]
    })

Genders_ordered = Genders.reindex (columns=['Gender','Percentage of Players','Count of Gender'])

Genders_ordered
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Normalized Totals

purchase_gender = purchase_data[["Gender","Price"]]
groups = purchase_gender.groupby("Gender")

gender_analysis["Average Purchase Price"] = round(groups[["Price"]].mean(),2)
gender_analysis["Purchase Count"] = groups[["Price"]].count()
gender_analysis["Total Purchase Value"] = groups[["Price"]].sum()
gender_analysis["Normalized Total"] = gender_analysis["Purchase Count"] / len(purchase_data["Gender"])

gender_analysis = gender_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Normalized Total"]]
gender_analysis
```

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age = purchase_data[["Age","Price"]]
age["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
age.head()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Normalized Totals

age_groups = age.groupby("Age Group")
age_demographics["Purchase Count"] = age_groups[["Price"]].count()
age_demographics["Average Purchase Price"] = round(age_groups[["Price"]].mean(),2)
age_demographics["Total Purchase Value"] = age_groups[["Price"]].sum()
age_demographics["Normalized Total"] = age_demographics["Purchase Count"] / len(purchase_data["Age"])

age_demographics = age_demographics[["Purchase Count","Average Purchase Price","Total Purchase Value", "Normalized Total"]]
age_demographics
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value

popular_items = purchase_data[["Item ID", "Item Name","Price"]]
popular_items = popular_items.groupby(["Item ID", "Item Name"])

item_popularity["Purchase Count"] = popular_items[["Price"]].count()
item_popularity["Item Price"]= round(popular_items[["Price"]].mean(),2)
item_popularity["Total Purchase Value"] = popular_items[["Price"]].sum()


item_popularity= item_popularity[["Purchase Count", "Item Price", "Total Purchase Value"]]
item_popularity = item_popularity.sort_values(["Purchase Count"],ascending = False)
item_popularity.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_count = purchase_data.groupby(['Item Name']).count().reset_index()
item_count = item_count.drop(columns=['Age','Gender','Item ID','SN'])
item_price = purchase_data.groupby(['Item Name']).mean().reset_index()
item_price = item_price.drop(columns=['Age','Item ID'])
item_value = purchase_data.groupby(['Item Name']).sum().reset_index()
item_value = item_value.drop(columns=['Age','Item ID'])


final = item_count
final['Item Price'] = item_price['Price']
final['Total Purchase Value'] = item_value['Price']
final.sort_values(['Price'], ascending=False).head().style.format({'Item Price':'${:,.2f}','Total Purchase Value':'${:,.2f}'})
```
